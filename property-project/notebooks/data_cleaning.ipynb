# Data Cleaning

```python
from __future__ import print_function  # Compatability with Python 3
print( 'Print function ready to serve.' )

import pandas as pd
pd.set_option('display.max_columns', 150)

import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv('data/property_listings_dataset.csv', low_memory=False)
```

```python
df.shape
```

```python
# Drop duplicates by listing_id
print( df.shape )
df = df.drop_duplicates(subset='listing_id')
print( df.shape )
sns.countplot(y='county', data=df)
```

```python
# Drop observations we are not interested it
df.shape
df = df[df.property_type != 'Land']
df = df[df.property_type != 'Parking/garage']
df = df[df.property_type != 'Barn conversion']
df = df[df.property_type != 'Office']
df = df[df.property_type != 'Farm']
df = df[df.property_type != 'Equestrian property']
df = df[df.property_type != 'Mobile/park home']
df = df[df.property_type != 'Retail premises']
df = df[df.property_type != 'Block of flats']
df = df[df.property_type != 'Hotel/guest house']
df.shape
```

```python
# Join bungalows as they are sparse classes into 1
df.property_type.replace(['Terraced bungalow','Detached bungalow', 'Semi-detached bungalow'], 'Bungalow', inplace=True)
sns.countplot(y='property_type', data=df)
```

# Labelling missing data
Now we are going to address missing categorical data. 

First, find and count missing data in observations

```python
df.select_dtypes(include=['object']).isnull().sum()
```

```python
df.to_csv('data/backup.csv', index=None)
```

```python
for column in df.select_dtypes(include=['object']):
    df[column] = df[column].fillna('Missing')
```

```python
df.select_dtypes(include=['object']).isnull().sum()
```

```python
df.to_csv('data/cleaned_with_features.csv', index=None)
```
