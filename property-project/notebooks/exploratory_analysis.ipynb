```python
# Bar plots
sns.countplot(y='property_type', data=df)
plt.show()
sns.countplot(y='status', data=df)
plt.show()
sns.countplot(y='category', data=df)
plt.show()
sns.countplot(y='price_modifier', data=df)
plt.show()
sns.countplot(y='price_change_summary.direction', data=df)
plt.show()
```

## Segmentations

```python
# Segment price by property_type and plot distributions
sns.boxplot(y='property_type', x='price', data=df)
plt.show()
sns.boxplot(y='property_type', x='num_bedrooms', data=df)
plt.show()
```

```python
# Segment by property_type and display the means within each class
df.groupby('property_type').mean()
```

```python
# Segment by property_type and display the means and standard deviations within each class
df.groupby('property_type').agg(['mean', 'std'])
```

```python
# Calculate correlations between numeric features
correlations = df.corr()
correlations
```

```python
# Make the figsize 7 x 6
plt.figure(figsize=(7,6))
# Plot heatmap of correlations
sns.heatmap(correlations)
```

```python
# Change color scheme
sns.set_style('white')

# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Make the figsize 9 x 8
plt.figure(figsize=(9,8))
# Plot heatmap of correlations
sns.heatmap(correlations*100, annot=True, mask=mask, cbar=False)
# Plot heatmap of correlations
```
